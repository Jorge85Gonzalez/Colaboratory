```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num = 20
# Two dimentionals arrays
data = np.random.rand(4, num)
print(data)
```

```python
plt.scatter(data[0], data[1], c=data[2], s=data[3]*1000, alpha=0.3, cmap='viridis')
plt.colorbar()
```
